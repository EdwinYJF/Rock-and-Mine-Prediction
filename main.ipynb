```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#separating data and labels
X = df.drop(60, axis=1)
y = df[60]
```

Training and test data split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)
```

Cross validation

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
```

```python
model = LogisticRegression(random_state=1)
scores = cross_val_score(estimator=model, X=X_train, y=y_train, scoring="accuracy", cv=skf)
```

```python
scores
```

```python
scores.mean()
```

Model training

```python
model.fit(X_train, y_train)
```

Model evaluation

```python
#training accuracy
y_pred = model.predict(X_train)
```

```python
accuracy = accuracy_score(y_train, y_pred)
accuracy
#the training accuracy is 83%
```

```python
#testing accuracy
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
#the testing accuracy is 76%
```

Hyperparameter tuning with cross_val_score

```python
C_values = np.arange(0.01, 1.1, 0.1) #the c values range from 0.01 to 1.0 with steps of 0.1
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)

for c in C_values:
    model = LogisticRegression(C=c, max_iter=1000, random_state=1)
    scores = cross_val_score(estimator=model, X=X_train, y=y_train, scoring="accuracy", cv=skf)
    print(f"C={c:.2f}  |  Mean CV Accuracy: {scores.mean():.4f}")
```

Comprehensive hyperparameter tuning with GridSearchCV

```python
params_grid = {
    "C": np.logspace(-3,2,5),
    "penalty": ["elasticnet"],
    "solver": ["saga"],
    "class_weight": [None, "balanced"],
    "l1_ratio": np.arange(0.0, 1.1, 0.1)
}

model = LogisticRegression(max_iter=1000, random_state=1)

grid_search = GridSearchCV(estimator=model, param_grid=params_grid, scoring="accuracy", cv=skf, n_jobs=-1, error_score="raise")

grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_score_)
```

```python
print(grid_search.best_params_)
```
